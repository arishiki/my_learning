```python
import numpy as np

np.random.randint(1,10,size=5)
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a grid graph
rows, cols = 10, 10
G = nx.grid_2d_graph(rows, cols)
weights = np.random.randint(1,10,size=len(G.edges()))

l1 = []

# Add weights to the edges
for i, (u, v) in enumerate(G.edges()):
    G.edges[u, v]['weight'] = weights[i]
    l1.append(((u,v), weights[i]))

sorted_list = sorted(l1, key = lambda x:x[1])
```

```python
pos = pos = {(x, y): (x, -y) for x, y in G.nodes()}

# Draw the graph
plt.figure(figsize=(8, 8))
nx.draw(G, pos, with_labels=False, node_size=50, node_color='black')

# Draw edge labels
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

# Show plot
plt.axis('off')
plt.show()
```

```python
edge_list = []
for edge, weight in sorted_list:
    edge_list.append(edge)
    try:
        nx.find_cycle(G.edge_subgraph(edge_list),source=edge[0], orientation=edge[1])
    except:
        pass
    else:
        del edge_list[-1]
print(edge_list)
```

```python
H = G.edge_subgraph(edge_list)
print(H.edges())
pos = {(x, y): (x, -y) for x, y in H.nodes()}

# Draw the graph
plt.figure(figsize=(8, 8))
nx.draw(H, pos, with_labels=False, node_size=50, node_color='black')

# Draw edge labels
edge_labels = nx.get_edge_attributes(H, 'weight')
nx.draw_networkx_edge_labels(H, pos, edge_labels=edge_labels)

# Show plot
plt.axis('off')
plt.show()
```
